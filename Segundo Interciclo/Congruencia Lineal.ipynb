```python
#Importamos librerias
import matplotlib.pyplot as plt
import pandas as pd
import psutil
```

<b> - Método que calculara la congruencia lineal y me retorna una lista de números en base a la a una semilla y la cantidad de números medios que deseo sacar. </b>

```python
def congruencia(semilla,iteraciones,a,c,m, dig):
    lista = []
    for i in range(1, iteraciones):
        xn = (a*semilla + c) % m
        semilla = xn
        lista.append(xn)
    return lista
```

* Semillas 

```python
#aleatorias
semilla1 = 20202020
semilla2 = 20212021
semilla3 = 19970827
semilla4 = 20180805
semilla5 = 19961207
semilla6 = psutil.virtual_memory().total #memoria virtual
semilla7 = int(psutil.cpu_freq().current) #Frecuencias d HZ
semilla8 = int(psutil.disk_usage('/').total/10000000) #Capasidad disco local C
semilla9 = psutil.disk_io_counters().write_count #Número de lecturas del disco
semilla10 = psutil.net_io_counters().bytes_sent #Numero de paquetes recibidos
#Lista de semillas
semillas = [semilla1,semilla2,semilla3,semilla4,semilla5,semilla6,semilla7,semilla8,semilla9,semilla10]
```

```python
cantidad = 100
num_digi = 5
a=semillas[5]
c=semillas[8]
M=19 
d = {}

#calculamos la lista de numeros randomicos 
for l in semillas:
    lista = congruencia(l, cantidad,a,c,M,num_digi)
    #llenamos un dataFrame con las listas
    d[str(l)] = lista
#Mostramos DF | lista de randomicos    
df = pd.DataFrame(data=d)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)
```
